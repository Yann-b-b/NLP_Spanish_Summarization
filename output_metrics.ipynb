```python
import pandas as pd
import matplotlib.pyplot as plt

# Yann, here is a script that creates and saves bar plots for each metric:

data = {
    "Model": ["GPT-Neo-125M", "Phi-1.5B", "TinyLlama-1.1B"],
    "BLEU": [0.31523055069185624, 0.06314816616579609, 0.07037502738216501],
    "METEOR": [0.47421366209585336, 0.33234296060530033, 0.3400551903876177],
    "ROUGE-1": [0.6728493000605318, 0.5053761684981459, 0.513228385239058],
    "ROUGE-2": [0.6347015909692957, 0.4664877422719237, 0.48016720016845493],
    "ROUGE-L": [0.6619365773900014, 0.49473945633886984, 0.5034278855770309]
}

df = pd.DataFrame(data).set_index("Model")

# Generate and save bar plots for each metric
for metric in ["BLEU", "METEOR", "ROUGE-1", "ROUGE-2", "ROUGE-L"]:
    plt.figure()
    df[metric].plot(kind='bar')
    plt.title(f"{metric} Scores")
    plt.xlabel("Model")
    plt.ylabel("Score")
    plt.tight_layout()
    plt.savefig(f"{metric}_scores.png")  # Save each plot as an image
    plt.show()
```
