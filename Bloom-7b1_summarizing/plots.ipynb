```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
file_paths = [
    "graded_biographies_part1_bloom.csv",
    "graded_biographies_part2_bloom.csv",
    "graded_biographies_part3_llama2.csv"
]

# Combine data from all files
all_grades = []

for file in file_paths:
    df = pd.read_csv(file)
    all_grades.extend(df["grade"].tolist())

# Convert to DataFrame
grades_df = pd.DataFrame(all_grades, columns=["grade"])

# Count occurrences of each grade
grade_counts = grades_df["grade"].value_counts(normalize=True).sort_index() * 100  # Convert to percentage

# Plot
plt.figure(figsize=(8, 5))
plt.bar(grade_counts.index, grade_counts.values, color='blue', alpha=0.7)
plt.xlabel("Grade (1 to 5)")
plt.ylabel("Percentage of Total Grades (%)")
plt.ylim(0, 100)
plt.title("Distribution of Grades Across All Datasets")
plt.xticks(range(1, 6))
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show plot

# Save the plot
plot_path = "grade_distribution.png"
plt.savefig(plot_path, dpi=300, bbox_inches="tight")
plt.show()

print(f"Graph saved as {plot_path}")

```
